```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import re
import string
from text_preprocessing import preprocess_text
```

```python
df1 = pd.read_csv("True.csv")
```

```python
df1.head()
```

```python
df2 = pd.read_csv("Fake.csv")
```

```python
df2.head()
```

```python
df2['text'][1]
```

```python
df1['target'] = "true" #add the target column as it is missing in the original df
```

```python
df2['target'] = "fake"
```

```python
data = [df1,df2] 
```

```python
df = pd.concat(data,ignore_index=True) #combine both the dataframes
"""
ignore index if not in df we'll have two rows with index 0,1,... if we ignore they
will be reassigned according to the size of new data frame
"""
```

```python
df.isnull().sum()  #no null values
```

```python
pd.DataFrame(df.groupby(['subject','target']).size())
#from this we can observe how much influence subject has on the target
```

```python
df1['subject'].unique()
```

```python
df2['subject'].unique()
```

```python
df['full_content'] = df['subject']+df['title']+df['text']
```

```python
df.head()
```

```python
df['processed_full_content'] = df['full_content'].apply(preprocess_text)
```

```python
df['processed_full_content'][0]
#no punctuations,stop words,all words are lemmatized it is completely processed now
```

```python
X = df['processed_full_content']
```

```python
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42,test_size=0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
v = TfidfVectorizer(max_features = 5000,ngram_range=(1,2))
```

```python
v.fit(X_train)
```

```python
X_train_tfidf = v.transform(X_train)
X_test_tfidf = v.transform(X_test)
```

```python
len(v.vocabulary_)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train_tfidf,y_train)
```

```python
y_pred = model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

```python
accuracy = accuracy_score(y_pred,y_test)
precision = precision_score(y_test, y_pred, average='binary', pos_label='true')
recall = recall_score(y_test, y_pred, average='binary', pos_label='true')
f1 = f1_score(y_test, y_pred, average='binary', pos_label='true')
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision (for positive class): {precision:.4f}")
print(f"Recall (for positive class): {recall:.4f}")
print(f"F1-Score (for positive class): {f1:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import joblib
```

```python
joblib.dump(model,'naive_bayes_model.pkl')
print("Trained Naive Bayes model saved as 'naive_bayes_model.pkl'")
```

```python
joblib.dump(v, 'tfidf_vectorizer.pkl')
print("Fitted TF-IDF vectorizer saved as 'tfidf_vectorizer.pkl'")
```
